```python
import pandas as pd
```

# Данные

Использованы данные роста-веса случайной выборки людей.

```python
data = pd.read_csv('weight-height.csv')
```

```python
data
```

```python
data.plot(y = 'Height', x = 'Weight', kind = 'scatter')
```

# Построение модели

```python
from sklearn.linear_model import LinearRegression
```

```python
X = pd.DataFrame(data['Weight'])
y = pd.DataFrame(data['Height'])

reg = LinearRegression().fit(X = X, y = y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, reg.predict(X), color = 'g')

plt.show()
```

```python
plt.scatter(X, y - reg.predict(X))
```

```python
pd.DataFrame(zip(X.columns, reg.coef_))
```

```python
reg.intercept_
```

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from mpl_toolkits import mplot3d
```

# Данные

Были использованы искусственно сгенерированные двухмерные данные.

```python
df = pd.DataFrame(np.load('clusterable_data.npy'),
                          columns=['x', 'y'])
```

```python
df.plot(kind='scatter',
               x='x',
               y='y',
               figsize=(12, 8))
```

```python
from sklearn.neighbors import NearestNeighbors
from matplotlib import pyplot as plt

neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(df)
distances, indices = neighbors_fit.kneighbors(df)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

```python
dbscan = DBSCAN(eps=0.033,
               min_samples=20,
               algorithm='brute',
               p=2)
preds = dbscan.fit_predict(df)
```

```python
df.plot(kind='scatter', x='x', y='y', c=preds, cmap='Accent_r', figsize=(12,8))
```
